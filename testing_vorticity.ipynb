```python
import numpy as np
import matplotlib.pyplot as plt
import Physical_Quantities.vorticity as vorticity
```

## Testing Vorticity on Example

```python
x_p = np.linspace(-35, 35, num=28)
y_p = np.linspace(-35, 35, num=28)
X_p, Y_p = np.meshgrid(x_p,y_p)
h=4
exp_data_fake = {}

exp_data_fake['x'] = X_p.flatten()[None,:]
exp_data_fake['y'] = Y_p.flatten()[None,:]

exp_data_fake['vx'] = -Y_p.flatten()[None,:] / np.sqrt((X_p.flatten()[None,:]**2+Y_p.flatten()[None,:]**2 + 1e-5))
exp_data_fake['vy'] = X_p.flatten()[None,:] / np.sqrt((X_p.flatten()[None,:]**2+Y_p.flatten()[None,:]**2+1e-5))

vorticity_grid = vorticity.get_vorticity(exp_data_fake, x_p, y_p, h, 70)
```

```python
plt.quiver(exp_data_fake['x'],exp_data_fake['y'],exp_data_fake['vx'],exp_data_fake['vy'])
```

```python
plot = plt.pcolormesh(X_p,Y_p, vorticity_grid[0])
plt.colorbar(plot)
plt.show()
```
